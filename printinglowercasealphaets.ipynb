```python
char = input("Enter a lowercase alphabet: ")
if char == "a":
    i=1
    p="*"
    for i in range(6):
        if  i==1:
            print(" "*3,p*2)
        elif i==2:
            print(" "*5,p)
        elif i==3:
            print("",p*6)
        elif i==4:
            print(p*1," "*3,p*1)
        elif i==5:
            print("",p*5,p)
            
elif char=="b":
    i=1
    p="*"
    for i in range(7):
        if  i==1:
            print(p)
        elif i==2:
            print(p)
        elif i==3:
            print(p*4)
        elif i==4:
            print(p," ",p)
        elif i==5:
            print(p*4)

elif char=="c":
    i = 1
    p = "*"
    for i in range(7):
        if  i==1:
            print("",p*3)
        elif i==3:
            print(p*1," "*4)
        elif i==5:
            print("",p*3)
            
elif char == "d":
    i = 1
    p = "*"
    for i in range(6):
        if  i==1:
            print(" "*3,p)
        elif i==2:
            print(" "*3,p)
        elif i==3:
            print("",p*4)
        elif i==4:
            print(p," ",p)
        elif i==5:
            print("",p*3,p)
            
elif char == "e":
    i = 1
    p = "*"
    for i in range(7):
        if  i==1:
            print("",p*3)
        elif i==2:
            print(p,p*3)
        elif i==3:
            print(p)
        elif i==4:
            print("",p*3)
            
elif char=="f":
    i = 1
    p = "*"
    for i in range(6):
        if  i==1:
            print(" ",p*3)
        elif i==2:
            print("",p)
        elif i==3:
            print(p*4)
        elif i==4:
            print("",p)
        elif i==5:
            print("",p)
            
elif char == "g":
    i = 1
    p = "*"
    for i in range(7):
        if  i==1:
            print("",p,p,p)
        elif i==2:
            print(p," ",p)
        elif i==3:
            print("",p,p*2)
        elif i==4:
            print(" "*3,p)
        elif i==5:
            print(p," "*1,p*1)
        elif i==6:
            print("",p*3,"")
            
elif char == "h":
    i = 1
    p = "*"
    for i in range(7):
        if  i==1:
            print(p*1)
        elif i==2:
            print(p*1)
        elif i==3:
            print(p*4)
        elif i==4:
            print(p*1," ",p*1)
        elif i==5:
            print(p*1," ",p*1)
            
elif char == "i":
    i = 1
    p = "*"
    for i in range(7):
        if  i==1:
            print(" ",p)
        elif i==2:
            print("",p*2)
        elif i==3:
            print(" ",p)
        elif i==4:
            print(p*5)
            
elif char == "j":
    i = 1
    p = "*"
    for i in range(7):
        if  i==1:
            print(" "*3,p)
            print(" ",p,p)
        elif i==2:
            print("   ",p*1)
        elif i==3:
            print(" "*3,p)
        elif i==4:
            print(p," ",p*1)
        elif i==5:
            print("",p,p)

elif char == "k":
    i = 1
    p = "*"
    for i in range(7):
        if  i==1:
            print(p)
        elif i==2:
            print(p,)
        elif i==3:
            print(p,"",p)
        elif i==4:
            print(p,p)
        elif i==5:
            print(p,"",p)
            
elif char == "l":
    i = 1
    p = "*"
    for i in range(7):
        if  i==1:
            print(p*2)
        elif i==2:
            print("",p)
        elif i==3:
            print("",p)
        elif i==4:
            print("",p)
        elif i==5:
            print(p*3)
            
elif char == "m":
    i = 1
    p = "*"
    for i in range(7):
        if  i==1:
            print(p," ",p," "*3,p)
        elif i==2:
            print(" ",p," ",p,p," ",p)
        elif i==3:
            print(" ",p," "*7,p)
            
elif char == "n":
    i = 1
    p = "*"
    for i in range(6):
        if  i==1:
            print(p," ",p*2)
        elif i==2:
            print(" ",p," ",p)
        elif i==3:
            print(" ",p," ",p)
            
elif char == "o":
    i = 1
    p = "*"
    for i in range(7):
        if  i==1:
            print("",p,p)
        elif i==2:
            print(p*1," ",p)
        elif i==5:
            print("",p,p)
            
elif char == "p":
    i = 1
    p = "*"
    for i in range(7):
        if  i==1:
            print(p,p*2)
        elif i==2:
            print("",p,"",p)
        elif i==3:
            print("",p*3)
        elif i==4:
            print("",p*1)
        elif i==5:
            print("",p)
            
elif char == "q":
    i = 1
    p = "*"
    for i in range(7):
        if  i==1:
            print("",p*4)
        elif i==2:
            print(p," "*2,p)
        elif i==3:
            print("",p*4,p)
        elif i==4:
            print(" "*5,p,"",p)
        elif i==5:
            print(" "*5,"",p*2)
        elif i==6:
            print(" "*6,p)
            
elif char == "r":
    i = 1
    p = "*"
    for i in range(7):
        if  i==1:
            print(p,p*3)
        elif i==2:
            print("",p,"",p*2)
        elif i==3:
            print("",p)
        elif i==4:
            print("",p)
            
elif char == "s":
    i = 1
    p = "*"
    for i in range(7):
        if  i==1:
            print("",p*5)
        elif i==2:
            print(p*1," ")
        elif i==3:
            print("",p*5)
        elif i==4:
            print(" "*5,p*1)
        elif i==5:
            print(p*5)
            
elif char == "t":
    i = 1
    p = "*"
    for i in range(7):
        if  i==1:
            print(" ",p)
        elif i==2:
            print("",p*3)
        elif i==3:
            print(" ",p*1)
        elif i==4:
            print(" ",p,"",)
        elif i==5:
            print("  ",p*2)
            
elif char == "u":
    i = 1
    p = "*"
    for i in range(5):
        if  i==1:
            print(p*1," ",p*1)
        elif i==2:
            print(p*1," ",p*1)
        elif i==3:
            print("",p*3,p)
            
elif char == "v":
    i = 1
    p = "*"
    for i in range(5):
        if  i==1:
            print(p*1," "*3,p*1)
        elif i==2:
            print("",p," ",p)
        elif i==3:
            print(" "*2,p)
            
elif char == "w":
    i = 1
    p = "*"
    for i in range(5):
        if  i==1:
            print(" ",p," "*5,p)
        elif i==2:
            print(" ",p,"",p,p,"",p)
        elif i==3:
            print(" "*2,p," "*3,p)
            
elif char == "x":
    i = 1
    p = "*"
    for i in range(5):
        if  i==1:
            print(" ",p," ",p)
        elif i==2:
            print(" "*3,p)
        elif i==3:
            print(" ",p," ",p)
            
elif char == "y":
    i = 1
    p = "*"
    for i in range(8):
        if i==1:
            print(p," "*3,p)
        elif i==2:
            print(p," "*3,p)
        elif i==3:
            print("",p,p,p)
        elif i==4:
            print(" "*4,p)
        elif i==5:
            print(" "*4,p)
        elif i==6:
            print(p," "*2,p)
        elif i==7:
            print(" "*1,p,p,"")
            
elif char == "z":
    i = 1
    p = "*"
    for i in range(5):
        if  i==1:
            print("",p*4)
        elif i==2:
            print(" "*3,p)
        elif i==3:
            print(" "*1,p)
        elif i==4:
            print("",p*4)
            
else:
    print("Your input isn't an alphabet")
```
